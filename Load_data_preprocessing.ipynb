# Preprocessing
****Handling Missing Data present in dataset****

```python
df_handled_missingData = df.fillna(value=df.mean())
df_handled_missingData.head()
```

**Handling categorical data**

```python
df_handled_missingData['Location'] = df_handled_missingData['Location'].map({'Mysore':1,'Mandya':2,'Raichur':3,'Koppal':4})
df_catog = df_handled_missingData
df_catog.head()
```

**Normalization**

```python
df_crop = df_catog.iloc[:,1:9]
df_crop = (df_crop - df_crop.mean()) / (df_crop.max() - df_crop.min())
print("\n*****Normalized Data*****\n")
print(df_crop.head())
```
